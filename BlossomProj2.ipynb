```python
#Modeling 
#Importing libraries
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import train_test_split
```

```python
#Multiple Regression without Engineered Features
#Loading default features
Training_data_mul=Train.merge(Riders, on='Rider Id')
##Dropping irrelevant features
Training_data_mul.drop('Order No', axis=1, inplace=True)
Training_data_mul.drop('User Id', axis=1, inplace=True)
Training_data_mul.drop('Vehicle Type', axis=1, inplace=True)
Training_data_mul.drop('Rider Id', axis=1, inplace=True)
Training_data_mul.drop('Arrival at Destination - Day of Month', axis=1, inplace=True)
Training_data_mul.drop('Arrival at Destination - Weekday (Mo = 1)', axis=1, inplace=True)
Training_data_mul.drop('Arrival at Destination - Time', axis=1, inplace=True)
Training_data_mul.drop('Placement - Time', axis=1, inplace=True)
Training_data_mul.drop('Confirmation - Time', axis=1, inplace=True)
Training_data_mul.drop('Arrival at Pickup - Time', axis=1, inplace=True)
Training_data_mul.drop('Pickup - Time', axis=1, inplace=True)
Training_data_mul.drop('Precipitation in millimeters', axis=1, inplace=True)
Training_data_mul.drop('Personal or Business', axis=1, inplace=True)
#Imputing missing data in Temperature column - (Because it has less missing data (20.59%))
Training_data_mul['Temperature']=Training_data_mul['Temperature'].fillna(int(Training_data_mul['Temperature'].median()))
#Defining dependent and independent features
target_column_mul = ['Time from Pickup to Arrival'] 
predictors_mul = list(set(list(Training_data_mul.columns))-set(target_column_mul))
#Normalizing independent data
Training_data_mul[predictors_mul] = Training_data_mul[predictors_mul]/Training_data_mul[predictors_mul].max()
#Creating the Training and Test Datasets
X_t = Training_data_mul[predictors_mul].values
y_t = Training_data_mul[target_column_mul].values
X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(X_t, y_t, test_size=0.20, random_state=40)
#Multiple Regression
lr_mul = LinearRegression()
lr_mul.fit(X_train_t, y_train_t)
#Predicting on test data
pred_test_lr_mul= lr_mul.predict(X_test_t)
#Model Evaluation with rmse
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
rmse_val = rmse(np.array(y_test_t), np.array(pred_test_lr_mul))
#Model Evaluation with mse
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_test_t, pred_test_lr_mul)
#Model Evaluation with mape
def mean_absolute_percent_error(y_true, y_pred):
   return np.mean(np.abs((y_true - y_pred)/y_true))*100
mape=mean_absolute_percent_error(y_test_t,pred_test_lr_mul)
#Model Evaluation with R-Squared
from sklearn.metrics import r2_score
r2=r2_score(y_test_t, pred_test_lr_mul)*100
method_name = 'Multiple Regression without Engineered Features'
print('Method: %s' %method_name)
print('RMSE: %.4f' %rmse_val)
print('MSE: %.4f' %MSE)
#print('MAPE: %.4f' %mape)
#print('R-Square: %.4f' %r2)
```

```python
#Multiple Regression with Engineered Features
#Creating Arrays for the Features and the Response Variable
target_column = ['Time from Pickup to Arrival'] 
predictors = list(set(list(Training_data.columns))-set(target_column))
#Normalizing independent data
Training_data[predictors] = Training_data[predictors]/Training_data[predictors].max()
#Creating the Training and Test Datasets
X = Training_data[predictors].values
y = Training_data[target_column].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=40)
#Multiple Regression model
lr = LinearRegression()
lr.fit(X_train, y_train)
#Predicting on test data
pred_test_lr= lr.predict(X_test)
#Model Evaluation with rmse
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
rmse_val = rmse(np.array(y_test), np.array(pred_test_lr))
rmse_val
#Model Evaluation with mse
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_test, pred_test_lr)
#Model Evaluation with mape
def mean_absolute_percent_error(y_true, y_pred):
   return np.mean(np.abs((y_true - y_pred)/y_true))*100
mape=mean_absolute_percent_error(y_test,pred_test_lr)
mape
#Model Evaluation with R-Squared
from sklearn.metrics import r2_score
r2=r2_score(y_test, pred_test_lr)*100
method_name = 'Multiple Regression with Engineered Features'
print('Method: %s' %method_name)
print('RMSE: %.4f' %rmse_val)
print('MSE: %.4f' %MSE)
#print('MAPE: %.4f' %mape)
#print('R-Square: %.4f' %r2)
```
